### Task 2: Getting Class Statistics
1. Create a function named `get_grades()` for computing the semestral grade of each student:

  `inputs`: `dictionary` of a class grade sheet.
  
  `outpus`: `dictionary` showing:
  * The prelim, midterms, and finals grades of each student
  * The semestral grade of each student

2. Create a function named `get_class_stats()`.
  
  `inputs`: `dictionary` from `get_grades()`.
  
  `outpus`: `dictionary` showing:
  * The lowest and highest prelim, midterm, finals, and semestral grades
  * the mean, median, mode, and standard deviation of the grades


```python
def get_grades(df_dict): 
  prelims = df_dict.iloc[:,3]
  midterms = df_dict.iloc[:,4]
  finals = df_dict.iloc[:,5]
  for semgrade in df_dict:
    semgrade=(prelims * 0.3) + (midterms * 0.3) + (finals * 0.4)
  return semgrade
```

```python
df_dict['semestral_grades'] = get_grades(df_dict)
df_dict.round(2)

```

```python
def get_class_stats(df_dict):
  pGrade_lowest = df_dict.iloc[:,3].min()
  pGrade_highest = df_dict.iloc[:,3].max()
  pGrade_mean = df_dict.iloc[:,3].mean()
  pGrade_median = df_dict.iloc[:,3].median()
  pGrade_mode = df_dict.iloc[:,3].mode()[0]
  pGrade_std = df_dict.iloc[:,3].std()

  mGrade_lowest = df_dict.iloc[:,4].min()
  mGrade_highest = df_dict.iloc[:,4].max()
  mGrade_mean = df_dict.iloc[:,4].mean()
  mGrade_meadian = df_dict.iloc[:,4].median()
  mGrade_mode = df_dict.iloc[:,4].mode()[0]
  mGrade_std = df_dict.iloc[:,4].std()

  fGrade_lowest = df_dict.iloc[:,5].min()
  fGrade_highest = df_dict.iloc[:,5].max()
  fGrade_mean = df_dict.iloc[:,5].mean()
  fGrade_median = df_dict.iloc[:,5].median()
  fGrade_mode = df_dict.iloc[:,5].mode()[0]
  fGrade_std = df_dict.iloc[:,5].std()

  sGrade_lowest = df_dict.iloc[:,6].min()
  sGrade_highest = df_dict.iloc[:,6].max()
  sGrade_mean = df_dict.iloc[:,6].mean()
  sGrade_median = df_dict.iloc[:,6].median()
  sGrade_mode = df_dict.iloc[:,6].mode()[0]
  sGrade_std = df_dict.iloc[:,6].std()
  
  
  class_stats = {'':["prelim_grades", "midterm_grades", "final_grades","semestral_grades"], 
                 'Lowest':[pGrade_lowest,mGrade_lowest,fGrade_lowest, sGrade_lowest], 
                 'Highest':[pGrade_highest,mGrade_highest,fGrade_highest,sGrade_highest], 
                 'Mean':[pGrade_mean,mGrade_mean, fGrade_mean, sGrade_mean], 
                 'Median':[pGrade_median,mGrade_meadian, fGrade_median, sGrade_median], 
                 'Mode':[pGrade_mode,mGrade_mode,fGrade_mode,sGrade_mode],
                 'Standard Deviation':[pGrade_std, mGrade_std, fGrade_std, sGrade_std]}
  df_dict = pd.DataFrame(class_stats)
  return df_dict
```
